```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/datasets/train-data.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.drop(['Unnamed: 0','New_Price'],axis=1,inplace=True)
df
```

```python
#removing the units in cols
```

```python
df['Mileage']=df['Mileage'].str.replace('km/kg','')
df['Mileage']=df['Mileage'].str.replace('kmpl','')
df
```

```python
df['Engine']=df['Engine'].str.replace('CC','')
df['Power']=df['Power'].str.replace('bhp','')
df
```

```python
df.dtypes
```

```python
#coverting mileage,engine,power to numerical type
```

```python
df['Mileage']=df['Mileage'].astype('float')
```

```python
df['Engine']=df['Engine'].astype('float')
```

```python
df.dtypes
```

```python
#example
```

```python
x='3.4'
type(x)
```

```python
y=float(x)
type(y)
```

```python
df['Power']=df['Power'].str.replace('null','0')
```

```python
df['Power']=df['Power'].astype('float')
```

```python
df['Power'].value_counts()
```

```python
df.loc[df['Power']=='null ']
```

```python
df.dtypes
```

```python
#filling the missing values
```

```python
df['Mileage']=df['Mileage'].fillna(df['Mileage'].mean())
```

```python
df['Power']=df['Power'].fillna(df['Power'].mean())
```

```python
df['Engine']=df['Engine'].fillna(df['Engine'].mode()[0])
```

```python
df['Seats']=df['Seats'].fillna(df['Seats'].mode()[0])
```

```python
df.isna().sum()
```

```python
#adding a new col brand from name col,
```

```python
#example
name1='Maruti Wagon R LXI CNG'
name1.split()[0]
```

```python
df['Brand']=df['Name'].apply(lambda x:x.split()[0])
df
```

```python
#droping the col,name
df.drop(['Name'],axis=1,inplace=True)
```

```python
#changing object into numeric values
```

```python
df['Fuel_Type'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Fuel_Type']=le.fit_transform(df['Fuel_Type'])
df
```

```python
#to get back the old df,we use le.inverse_transform(df['fueltype'])
```

```python
df['Brand'].value_counts()
```

```python
df['Transmission']=le.fit_transform(df['Transmission'])
df['Owner_Type']=le.fit_transform(df['Owner_Type'])
df['Brand']=le.fit_transform(df['Brand'])
df['Location']=le.fit_transform(df['Location'])
```

```python
df['Location'].value_counts()
```

```python
#multiple regression
```

```python
X=df.drop(['Price'],axis=1)
X
```

```python
y=df['Price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
```

```python
y_pred=mlr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print('r2_score=',r2_score(y_test,y_pred))
```

```python
print('MAE=',mean_absolute_error(y_test,y_pred))
```

```python
print('MSE',mean_squared_error(y_test,y_pred))
```

```python
import numpy as np
```

```python
print('RMSE=',np.sqrt(y_test,y_pred))
```

```python
print('slopes=',mlr.coef_)
```

```python
print('y intercept=',mlr.intercept_)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(X_train,y_train)
```

```python
y_knn=knn.predict(X_test)
y_knn
```
